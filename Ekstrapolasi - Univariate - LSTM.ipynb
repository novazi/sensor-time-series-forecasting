## _Data Modelling_ untuk _Dataset_ Kelompok 1 Kecerdasan Buatan

## 5. Normalisasi Data dengan _Scaling_

Beberapa pemodelan machine learning seperti _neural network_ mengharuskan semua variabel memiliki rentang nilai yang sama, karena jika tidak sama, _feature_ dengan rentang nilai terbesar otomatis akan menjadi _feature_ yang paling mendominasi dalam proses _training_/komputasi, sehingga model yang dihasilkan pun akan sangat bias. Oleh karena itu, sebelum memulai _training model_, kita terlebih dahulu perlu melakukan _data rescaling_ ke dalam rentang 0 dan 1, sehingga semua _feature_ berada dalam rentang nilai tersebut, yaitu nilai max = 1 dan nilai min = 0.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_normal = pd.DataFrame(scaler.fit_transform(df_new[df_new.columns.difference(['Timestamp'])]), columns = df_new[df_new.columns.difference(['Timestamp'])].columns)
df_normal.head()
```

```python
df_newest = pd.merge(df_drop[['Timestamp']], df_normal, left_index=True, right_index=True)
df_newest
```

```python
df_newest.info()
```

```python
df_newest['Timestamp'] = pd.to_datetime(df_newest['Timestamp'])
df.set_axis(df_newest['Timestamp'], inplace=True)
```

<font size="4"> **Pemodelan**

Mempersiapkan modul/_library_ untuk persiapan pemodelan.

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, RepeatVector, Flatten
from keras.layers import TimeDistributed
from math import sqrt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from matplotlib import pyplot
from pandas import concat
```

```python
close_data = df_newest['CapacityGenerator'].values
close_data = close_data.reshape((-1,1))

split_percent = 0.70
split = int(split_percent*len(close_data))

close_train = close_data[:split]
close_test = close_data[split:]
test = close_data[split:]

timestamp_train = df_newest['Timestamp'][:split]
timestamp_test = df_newest['Timestamp'][split:]

from keras.preprocessing.sequence import TimeseriesGenerator

look_back = 5
train_generator = TimeseriesGenerator(close_train, close_train, 
                                      length=look_back, batch_size=20)     
test_generator = TimeseriesGenerator(close_test, close_test, 
                                     length=look_back, batch_size=1)

model = Sequential()
model.add(
        LSTM(30,
        activation='relu',
        input_shape=(look_back,1)))
model.add(Dense(1,activation='relu'))
model.compile(optimizer='adam', loss='mse')

# fit network
history = model.fit_generator(train_generator, epochs=20, verbose=1)
```

```python
import plotly.graph_objects as go
import numpy as np

prediction = model.predict(test_generator)

close_train1 = close_train.reshape((-1))
close_test1 = close_test.reshape((-1))
prediction1 = prediction.reshape((-1))

trace1 = go.Scatter(
    x = timestamp_train,
    y = close_train1,
    mode = 'lines',
    name = 'Data'
)
trace2 = go.Scatter(
    x = timestamp_test,
    y = prediction1,
    mode = 'lines',
    name = 'Prediction'
)
trace3 = go.Scatter(
    x = timestamp_test,
    y = close_test1,
    mode='lines',
    name = 'Ground Truth'
)
layout = go.Layout(
    title = "Prediksi Data Test",
    xaxis = {'title' : "Timestamp"},
    yaxis = {'title' : "CapacityGenerator"}
)
fig = go.Figure(data=[trace1, trace2, trace3], layout=layout)
fig.show()
```
